```python
#Aqui criamos um novo dataframe apenas com as colunas que eu quero.
ca_df = combustiveis_df[['Revenda' , 'Municipio' ,'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
#Exibe a 4ª linha.
display(ca_df.loc[3])
```

```python
#Exibir da 10ª linha até a 20ª linha
display(ca_df.loc[9:19])
```

```python
# Criar um dataframe gas_df contendo
# Apenas as colunas (Revenda, Municipio, Produto, Valor de Venda)
# Somente com combustíveis sendo GASOLINA e exibir na tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir qual é o posto e o munícipio da gasolina mais cara do país?
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas condições para filtragem
#Quais são so preços do ETANOL na minha cidade (SALVADOR)
#Ordenado do menor valor de venda para o maior
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'SALVADOR')]
etanol_indaiatuba_df.sort_values(by='Valor de Venda', inplace=True)
display(etanol_indaiatuba_df)
```

```python
# Qual a média de preços dos combustíveis GASOLINA  e GASOLINA ADITIVADA do Bairro PERNAMBUES em SALVADOR?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'PERNAMBUES') & (combustiveis_df['Municipio'] == 'SALVADOR') & ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), ['Valor de Venda']].mean())
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro PERNAMBUES em SALVADOR? (Feito com isin)
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'PERNAMBUES') & 
                            (combustiveis_df['Municipio'] == 'SALVADOR') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL do Brasil todo?
media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```
