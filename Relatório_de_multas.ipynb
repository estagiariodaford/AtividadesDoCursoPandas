```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Multas_df = pd.read_csv("Multas_2007_1.csv", encoding='latin-1', sep=";" , on_bad_lines='skip')
display(Multas_df)
```

```python
tipodeabordagem_df = Multas_df["tip_abordagem"].value_counts().reset_index()

tipodeabordagem_df.rename(columns={'tip_abordagem':'Quantidade'}, inplace=True)
tipodeabordagem_df.rename(columns={'index':'Tipo de Abordagem'}, inplace=True)

display(tipodeabordagem_df)
```

```python
tipodeabordagem_df.plot(
    x='Tipo de Abordagem', 
    y='Quantidade', 
    kind='bar',
    title="Tabela de multa por quantidade de abordagem" 
)

plt.show()
```

```python
estadocommaiorinfracao_df = Multas_df["num_br_infracao"].value_counts().reset_index()

estadocommaiorinfracao_df.rename(columns={'num_br_infracao':'Quantidade'}, inplace=True)
estadocommaiorinfracao_df.rename(columns={'index':'Estado Multa'}, inplace=True)

display(estadocommaiorinfracao_df)
```

```python
estadocommaiorinfracao_df.plot(
    x='Estado Multa', 
    y='Quantidade', 
    kind='bar',
    title="Estados que registraram maior quantidade de multas em 2007.1" ,
    figsize=(14, 5)
)

plt.show()
```

```python
horarioinfracoes_df = Multas_df["cod_municipio_infracao"].value_counts().reset_index()

horarioinfracoes_df.rename(columns={'cod_municipio_infracao':'Quantidade'}, inplace=True)
horarioinfracoes_df.rename(columns={'index':'HorÃ¡rio das InfraÃ§Ãµes'}, inplace=True)
horarioinfracoes_df = horarioinfracoes_df.head(10)
horarioinfracoes_df.sort_values(by=['HorÃ¡rio das InfraÃ§Ãµes'], ascending=True, inplace=True)
display(horarioinfracoes_df)
```

```python
horarioinfracoes_df.plot(
    x='HorÃ¡rio das InfraÃ§Ãµes', 
    y='Quantidade', 
    kind='bar',
    title="Os 10 horÃ¡rios no qual foram registrados maior quantidade de multas em 2007.1" ,
    figsize=(14, 5)
)

plt.show()
```

```python
marcadainfracao_df = Multas_df["cod_veiculo_especie"].value_counts().reset_index()

marcadainfracao_df.rename(columns={'cod_veiculo_especie':'Quantidade'}, inplace=True)
marcadainfracao_df.rename(columns={'index':'Marca do Carro'}, inplace=True)
marcadainfracao_df = marcadainfracao_df.head(15)
display(marcadainfracao_df)
```

```python
marcadainfracao_df.plot(
    x='Marca do Carro', 
    y='Quantidade', 
    kind='bar',
    title="As 15 marcas de carros no qual foram registrados maior quantidade de multas em 2007.1" ,
    figsize=(14, 5)
)

plt.show()
```
